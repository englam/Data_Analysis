```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Amazon R7500v2 URL
url = "https://www.amazon.com/NETGEAR-Nighthawk-Ultimate-Gaming-Router/product-reviews/B01GPAOJJG/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&reviewerType=avp_only_reviews&filterByStar=one_star&pageNumber=1"

r7500v2 = requests.get(url)

#print (r7500v2.text)
```

```python
bsobj = BeautifulSoup(r7500v2.text)
#print (bsobj)
```

```python
''' First Bug'''
first_title = bsobj.find("", {"class": "a-size-base a-link-normal review-title a-color-base a-text-bold"}).text
first_content = bsobj.find("", {"class": "a-row review-data"}).text
first_time = bsobj.find("", {"class": "a-size-base a-color-secondary review-date"}).text
first_time = pd.to_datetime(first_time)
first_time = pd.Period(first_time, freq='D')
first_author = bsobj.find("", {"class": "a-size-base a-link-normal author"}).text

print ("bug title:",first_title,"\n")
print ("bug content:",first_content,"\n")
print ("bug time:",first_time)
print ("bug author:",first_author,"\n")

bugs_title_pd = pd.Series()
bugs_content_pd = pd.Series()
bugs_time_pd = pd.Series()
bugs_author_pd = pd.Series()

bugs_title_pd = bugs_title_pd.append(pd.Series([first_title]))
#bugs_title_pd = bugs_title_pd.append(pd.Series([first_title])).reset_index(drop=True)
bugs_content_pd = bugs_content_pd.append(pd.Series([first_content]))
bugs_time_pd = bugs_time_pd.append(pd.Series([first_time]))
bugs_author_pd = bugs_author_pd.append(pd.Series([first_author]))


print (bugs_title_pd)
print (bugs_content_pd)
print (bugs_time_pd)
print (bugs_author_pd)

#print (type(bugs_title))
```

```python
'''Other Bugs except first'''
for i in bsobj.find("", {"id": "cm_cr-review_list"}).find("", {"class": "a-section review"}).next_siblings:

    if i.find("", {"class": "a-size-base a-link-normal review-title a-color-base a-text-bold"}) is not None:
        bug_title = i.find("",{"class": "a-size-base a-link-normal review-title a-color-base a-text-bold"}).text

        bug_content = i.find("", {"class": "a-row review-data"}).text

        bug_time = i.find("", {"class": "a-size-base a-color-secondary review-date"}).text
        bug_time = pd.to_datetime(bug_time)
        bug_time = pd.Period(bug_time, freq='D')

        bug_author = i.find("", {"class": "a-size-base a-link-normal author"}).text
        bugs_title_pd = bugs_title_pd.append(pd.Series([bug_title])).reset_index(drop=True)
        print ("bug title:",bug_title)
        print ("bug time:",bug_time)
        print ("bug author:",bug_author)
        

        
        #bugs_content_pd = bugs_content_pd.append(pd.Series([bug_content])).reset_index(drop=True)
        #bugs_time_pd = bugs_time_pd.append(pd.Series([bug_time])).reset_index(drop=True)
        #bugs_author_pd = bugs_author_pd.append(pd.Series([bug_author])).reset_index(drop=True)
```
