## 分组计算

我们需要计算 30 个自然日里的股票平均波动周期。这样，我们必须以 30 天为单位，对所有的历史数据进行分组。然后逐个分组计算其波动率

```python
# 定义产生分组索引的函数，比如我们要计算的周期是 20 天，则按照日期，20 个交易日一组
def gen_item_group_index(total, group_len):
    """ generate an item group index array

    suppose total = 10, unitlen = 2, then we will return array [0 0 1 1 2 2 3 3 4 4]
    """

    group_count = total / group_len
    #print (round(group_count)) 不取小數點
    
    # example: gen_item_group_index(10, 3)
    # group_index[i * group_len: (i + 1) * group_len] = [0:3] , [3:6], [6:9]
    # group_index[(i + 1) * group_len : total] = group_index[9:10] = 3
    
    group_index = np.arange(total)
    for i in range(round(group_count)):
        group_index[i * group_len: (i + 1) * group_len] = i
    group_index[(i + 1) * group_len : total] = i + 1
    return group_index.tolist()

```

```python
gen_item_group_index(10, 3)
```

## 根据分组索引来分组

```python
period = 30

group_index = gen_item_group_index(len(data), period)
# 把分组索引数据添加到股票数据里
data['group_index'] = group_index
print (len(data))
data.head().append(data.tail())

# group index有分組成功了， 七月的前三十天為 0， 最後的12月是6
```

## 分组计算最高价和最低价

```python
print (data.loc['2008-07-02','ceiling_price'])
print (data.loc['2008-07-28','ceiling_price'])
print (data.loc['2008-08-13','ceiling_price'])
print (data.loc['2008-07-31','ceiling_price'])
print (data.loc['2008-09-18','ceiling_price'])
print (data.loc['2008-08-30','ceiling_price'])
```

```python
# 针对下跌的波动，我们把最高价设置为负数。什么是下跌的波动？就是先出现最高价，再出现最低价

# g.idxmin() 找出同一個group裡面 最低價 的日期
# g.idxmax() 找出同一個group裡面 最高價 的日期
# np.max(g) 找出同一個group裡面 最高價 的數值

def _ceiling_price(g):
    #print (g)
    print ("group , 最低價的日期",g.idxmin())
    print ("group , 最高價的日期",g.idxmax())
    print ("group , 最高價的數值",np.max(g))
    #print (-np.max(g))
    return g.idxmin() < g.idxmax() and np.max(g) or -np.max(g)


# 根据索引分组计算
group = data.groupby('group_index').agg({
                                        'volume': 'sum',
                                        'floor_price': 'min',
                                        'ceiling_price': _ceiling_price})
group.head()

```

```python
# 添加每个分组的起始日期
date_col = pd.DataFrame({"group_index": group_index, "date": idx})
group['date'] = date_col.groupby('group_index').agg('first')
group.head()

```

```python
# 添加我们的波动指标 股票波动系数 = 最高价/最低价
group['ripples_radio'] = group.ceiling_price / group.floor_price
group.head()
```

```python
# 降序排列。我们把分组的起始日期，交易量总和都列出来，也可以观察一下交易量和股票波动比的关系
ripples = group.sort_values('ripples_radio', ascending=False)
ripples.head()
```

```python
#计算前 5 大上涨波动的平均值
ripples.head(5).ripples_radio.mean()
```

```python
#计算前 10 大下跌波动的平均值
ripples.tail(5).ripples_radio.mean()
```
